```python
import plotly.express as px
import pandas as pd
import numpy as np
import csv
import sqlite3
import json
import plotly.graph_objects as go
#import dash
from dash import Dash, dcc, html, Input, Output
import jupyterlab_dash
```

```python
DBNAME = 't_h_readings.db'
con  = sqlite3.connect(DBNAME,
                      detect_types=sqlite3.PARSE_DECLTYPES |
                        sqlite3.PARSE_COLNAMES)

cur = con.cursor()
# con.row_factory = sqlite3.Row
#query = '''SELECT timestamp as "timestamp [timestamp]", temperature, humidity FROM reading LIMIT 500;'''
query = '''SELECT
                timestamp as "timestamp [timestamp]",
                temperature,
                avg(temperature) OVER (ORDER BY timestamp ROWS BETWEEN 10 PRECEDING AND 10 FOLLOWING) as wtemperature,
                humidity,
                avg(humidity) OVER (ORDER BY timestamp ROWS BETWEEN 20 PRECEDING AND 20 FOLLOWING) as whumidity
            FROM reading ORDER BY timestamp LIMIT 5000;'''
df = pd.read_sql_query(query, con)

fig = go.Figure([go.Scatter(x=df['timestamp'], y=df['wtemperature'])])
fig.show()
```

```python
fig = go.Figure([go.Scatter(x=df['timestamp'], y=df['whumidity'])])
fig.show()
```
